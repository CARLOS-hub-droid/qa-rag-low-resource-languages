# I - HuggingFace and Tokenisation

```python
# Step 1: Install HuggingFace transformers library
#!pip install transformers

# Step 2: Import the tokenizer
from transformers import AutoTokenizer

# Step 3: Load a multilingual model (supports 104 languages)
tokenizer = AutoTokenizer.from_pretrained(
            "bert-base-multilingual-cased"
)

# Step 4: Tokenize a simple sentence
text = "Hello, this is my first NLP code!"
tokens = tokenizer(text)

# Step 5: Print the result
print(tokens)
print("Number of tokens:", len(tokens['input_ids']))
```

# How to See the Vocabulary ourselves
#### We can decode the numbers back to words:

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")

# Your numbers
numbers = [101, 31178, 117, 10531, 10124, 15127, 10422, 81130, 11127, 16964, 106, 102]

# Decode numbers back to words
words = tokenizer.decode(numbers)
print(words)
```

# II - Multilingual Embeddings with LaBSE

```python
import os
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"
```

```python

# Step 1: Install sentence-transformers

#!pip install sentence-transformers

# Step 2: Import and load LaBSE model

from sentence_transformers import SentenceTransformer
model = SentenceTransformer("sentence-transformers/LaBSE")

# Step 3: Create embeddings for two sentences
sentences = ["Hello world", "Bonjour le monde"]
embeddings = model.encode(sentences)

# Step 4: Inspect the results
print("Embedding shape:", embeddings.shape)
print("First 5 values:", embeddings[0][:5])
print("Number of dimensions:", embeddings.shape[1])
```
